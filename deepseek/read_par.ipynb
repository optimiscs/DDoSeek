```python
import fastparquet
import pandas as pd
```

```python
import os
# 读取本地的 UDP-testing 的 parquet 文件
# file_path = 'UDP-testing.parquet'  # 替换为你的文件路径
file_paths = [
    'UDP-testing.parquet',
    # 'MSSQL-testing.parquet',
    # 'Syn-testing.parquet',
    # 'NetBIOS-testing.parquet',
]

for file_path in file_paths:
    # 提取文件名字，不保留后缀
    file_name = os.path.splitext(os.path.basename(file_path))[0]
    print(file_name)
    df = pd.read_parquet(f"DDoS-Defense/data/{file_path}", engine='fastparquet')

    # 将 Label 列转换为字符串类型
    df['Label'] = df['Label'].astype(str)

    import json


    # 将 DataFrame 转换为 JSON 格式
    json_data = []
    for _, row in df.iterrows():
        json_entry = {
            "instruction": "Bellow is a network packet which might be used as DDOS attack. Please classify the packet and give its type",
            "input": row.drop("Label").to_dict(),  # 将除 Label 外的列转换为字典
            "output": str(row["Label"]) if isinstance(row["Label"], str) else row["Label"]  # 确保 Label 是字符串或保持原始值
        }
        json_data.append(json_entry)

    # 保存为 JSON 文件
    output_file = f"{file_name}.json"
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(json_data, f, ensure_ascii=False, indent=4)

    print(f"JSON 数据已保存到 {output_file}")
    # # 查看前几行
    # print(df.head(1))

    # # 查看总行数
    # print(f"总行数: {len(df)}")
```
